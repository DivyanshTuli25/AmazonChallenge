```python
import tensorflow as tf
import tensorflow_hub as hub
import tensorflow_text as text
from sklearn.metrics.pairwise import cosine_similarity


from sklearn.model_selection import train_test_split
```

```python
encoder_url  = 'https://tfhub.dev/tensorflow/bert_en_uncased_L-12_H-768_A-12/4'
preprocess_url = 'https://tfhub.dev/tensorflow/bert_en_uncased_preprocess/3'
```

```python
bert_preprocess_model = hub.KerasLayer(preprocess_url)
bert_encoder = hub.KerasLayer(encoder_url)
```

```python
def get_embeddings(sentence):
  preprocessed_text = encoder_url(sentence)
  embedding = preprocess_url(preprocessed_text)['pooled_output']
```

```python
# Bert Layers
input_layer = tf.keras.layers.Input(shape = (),dtype = tf.string, name = 'Input')
preprocessed_text = bert_preprocess_model(input_layer)
output =  bert_encoder(preprocessed_text)
```

```python
output_1 = tf.keras.layers.Dropout(0.1, name = 'dropout')(output['pooled_output'])
output_cat = tf.keras.layers.Dense(1, activation = 'relu', name = 'category')(output_1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(Data_new_sample['DESCRIPTION'], Data_new_sample['PRODUCT_LENGTH'], test_size = 0.25, random_state = 5)
```

```python
model =tf.keras.Model(inputs = [input_layer], outputs = [output_cat])
```

```python
model.summary()
```

```python
tf.keras.metrics.MeanSquaredError(name="mean_squared_error", dtype=None)
```

```python
tf.keras.optimizers.Adam(
    learning_rate=0.1,
    name = 'adam'

)
```

```python
tf.keras.losses.MeanSquaredError(reduction="auto", name="mean_squared_error")
```

```python
model.compile(optimizer = 'adam', loss=tf.keras.losses.MeanSquaredError(), metrics =tf.keras.metrics.MeanSquaredError())
```

```python
model.fit(x_train, y_train, epochs =3, validation_data = [x_test, y_test])
```
